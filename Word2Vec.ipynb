```python
import string
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english')) - {"not", "no", "nor"}

def preprocess(text):
    # lowercase
    text = text.lower()
    
    # remove punctuation
    text = ''.join(c for c in text if c not in string.punctuation)
    
    # simple tokenization (split)
    words = text.split()
    
    # remove stopwords
    words = [w for w in words if w not in stop_words]
    
    return ' '.join(words)
```

# Train Word2Vec on training data

```python
from gensim.models import Word2Vec
import numpy as np

# -----------------------------
# 1. Train Word2Vec on TRAIN set only
# -----------------------------
w2v_model = Word2Vec(
    sentences=X_train_tokens,
    vector_size=100,
    window=5,
    min_count=2,
    workers=4,
    sg=1,          # 1 = Skip-gram, 0 = CBOW
    seed=42
)

# -----------------------------
# 2. Function to convert one document into one vector
#    using average of word vectors
# -----------------------------
def document_vector(tokens, model):
    valid_vectors = [model.wv[word] for word in tokens if word in model.wv]
    
    if len(valid_vectors) == 0:
        return np.zeros(model.vector_size)
    
    return np.mean(valid_vectors, axis=0)

# -----------------------------
# 3. Convert all train and test documents
# -----------------------------
X_train_w2v = np.array([document_vector(tokens, w2v_model) for tokens in X_train_tokens])
X_test_w2v = np.array([document_vector(tokens, w2v_model) for tokens in X_test_tokens])

# -----------------------------
# 4. Sanity checks
# -----------------------------
print("Train Word2Vec shape:", X_train_w2v.shape)
print("Test Word2Vec shape:", X_test_w2v.shape)
print("Word2Vec vector size:", w2v_model.vector_size)
print("Vocabulary size:", len(w2v_model.wv))

# -----------------------------
# 5. Check how many documents became all-zero vectors
# -----------------------------
train_zero_docs = np.sum(np.all(X_train_w2v == 0, axis=1))
test_zero_docs = np.sum(np.all(X_test_w2v == 0, axis=1))

print("Zero-vector documents in train set:", train_zero_docs)
print("Zero-vector documents in test set:", test_zero_docs)

# -----------------------------
# 6. Optional: inspect one sample
# -----------------------------
print("\nExample tokens:")
print(X_train_tokens.iloc[0])

print("\nExample document vector (first 10 values):")
print(X_train_w2v[0][:10])
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
import numpy as np

class Word2VecTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, w2v_model):
        self.w2v_model = w2v_model
        self.vector_size = w2v_model.vector_size

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return np.array([self.document_vector(text) for text in X])

    def document_vector(self, text):
        tokens = text.lower().split()
        vectors = [self.w2v_model.wv[word] for word in tokens if word in self.w2v_model.wv]

        if len(vectors) == 0:
            return np.zeros(self.vector_size)

        return np.mean(vectors, axis=0)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier

w2v_transformer = Word2VecTransformer(w2v_model)

pipeline_w2v_rf = Pipeline([
    ('w2v', w2v_transformer),
    ('rf', RandomForestClassifier(
        n_estimators=100,
        max_depth=None,
        min_samples_split=5,
        min_samples_leaf=1,
        random_state=42,
        n_jobs=-1
    ))
])

pipeline_w2v_rf.fit(X_train, y_train)
import pickle

with open("w2v_rf_pipeline.pkl", "wb") as f:
    pickle.dump(pipeline_w2v_rf, f)
```
