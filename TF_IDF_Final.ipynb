```python
import string
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english')) - {"not", "no", "nor"}

def preprocess(text):
    # lowercase
    text = text.lower()
    
    # remove punctuation
    text = ''.join(c for c in text if c not in string.punctuation)
    
    # simple tokenization (split)
    words = text.split()
    
    # remove stopwords
    words = [w for w in words if w not in stop_words]
    
    return ' '.join(words)
```

# Vectorization TF_IDF Revised

```python
custom_stopwords = {
    "dear", "sir", "madam", "regards", "sincerely",
    "kindly", "subject", "fw", "fwd", "re",
    "attached", "thank", "thanks", "hello"
}

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

all_stopwords = ENGLISH_STOP_WORDS.union(custom_stopwords)


vectorizer = TfidfVectorizer(
    #max_features=10000,
    max_features=8000,
    #min_df=5,
    min_df=10,
    #max_df=0.95,
    max_df=0.85,
    #ngram_range=(1,1) #Only single words,
    ngram_range=(1,2),
    token_pattern=r'\b[a-zA-Z]{2,}\b',
    sublinear_tf=True
)

X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
# TN (7818) → legit emails correctly classified
# FP (111) → legit → wrongly flagged as phishing ❗
# FN (550) → phishing → missed (VERY IMPORTANT ❗)
# TP (8019) → phishing correctly detected

# Class 0 = legit
# Precision 0.93 means 93% of emails predicted as legit are correct
# Recall 0.99 means 99% of actual legit emails are correctly identified

# Class 1 = phishing
# Precision 0.99 means 99% of emails predicted as phishing are correct
# Recall 0.93 means 93% of actual phishing emails are correctly identified

# F1-score is the harmonic mean of precision and recall, providing a single metric that balances both. 
# For class 0 (legit), an F1-score of 0.96 indicates a strong performance in correctly identifying legit emails while minimizing false positives. 
# For class 1 (phishing), an F1-score of 0.96 also indicates a strong performance in correctly identifying phishing emails while minimizing false negatives. 
# Overall, these metrics suggest that the model is performing well in distinguishing between legit and phishing emails.


#Class 0 → support = 7929
# Class 1 → support = 8569

# macro avg = (metric_class0 + metric_class1) / 2
# Macro precision = (0.93 + 0.99)/2 = 0.96
# Macro recall = (0.99 + 0.94)/2 = 0.96
# Macro F1-score = (0.96 + 0.96)/2 = 0.96

# weighted avg = (metric0 * support0 + metric1 * support1) / total
# Weighted precision = (0.93*7929 + 0.99*8569) / (7929+8569) = 0.96
# Weighted recall = (0.99*7929 + 0.94*8569) / (7929+8569) = 0.96
# Weighted F1-score = (0.96*7929 + 0.96*8569) / (7929+8569) = 0.96

```

# 2. Linear SVM

```python
from sklearn.svm import LinearSVC

svm_model = LinearSVC(random_state=42)
svm_model.fit(X_train_vec, y_train)

y_pred= svm_model.predict(X_test_vec)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVC
import pandas as pd

# Parameter grid
param_grid = {
    'C': [0.01, 0.1, 1, 10]
}

# GridSearch
grid_svm = GridSearchCV(
    LinearSVC(random_state=42, max_iter=2000),
    param_grid,
    cv=3,
    scoring='recall',
    n_jobs=-1,
    return_train_score=True
)

# Train
grid_svm.fit(X_train_vec, y_train)

# Best result
print("Best parameters:", grid_svm.best_params_)
print("Best recall:", grid_svm.best_score_)

# Convert results to DataFrame
df = pd.DataFrame(grid_svm.cv_results_)

# Expand params into separate columns
params_df = pd.json_normalize(df['params'])

# Combine with original results
df_full = pd.concat([params_df, df], axis=1)

# Select clean columns
df_results = df_full[['C', 'mean_train_score', 'mean_test_score', 'rank_test_score']]

# Sort by performance
df_results = df_results.sort_values(by='rank_test_score')

print("\nAll parameter combinations:")
print(df_results)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Get best model from GridSearch
best_model = grid_svm.best_estimator_

# Predict on test set
y_pred_svc= best_model.predict(X_test_vec)

# Evaluation
print(confusion_matrix(y_test, y_pred_svc))
print(classification_report(y_test, y_pred_svc))
```

# Model Comparison (Multinomial Naive Bayes, Linear SVC, Random Forest, Decision Tree, Logistic Regression)

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

# Put model names and predictions here
model_predictions = {
    'Multinomial Naive Bayes': y_pred_nb,
    'Linear SVC': y_pred_svc,
    'Random Forest': y_pred_rf,
    'Decision Tree': y_pred_dt,
    'Logistic Regression': y_pred_lr
}

results = []

for model_name, y_pred in model_predictions.items():
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    
    results.append({
        'Model': model_name,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision (Class 1)': precision_score(y_test, y_pred, pos_label=1),
        'Recall (Class 1)': recall_score(y_test, y_pred, pos_label=1),
        'F1-score (Class 1)': f1_score(y_test, y_pred, pos_label=1),
        'False Positives': fp,
        'False Negatives': fn
    })

comparison_df = pd.DataFrame(results)
comparison_df = comparison_df.sort_values(by='Recall (Class 1)', ascending=False)
comparison_df = comparison_df.round(2)

print(comparison_df)
```

```python
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
import pickle

#pipeline = Pipeline([
    #('tfidf', TfidfVectorizer()),
    #('model', LogisticRegression(
        #penalty='elasticnet',
        #solver='saga',
        #C=10,
        #l1_ratio=0.2,
        #max_iter=2000,
        #random_state=42
    #))
#])

pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(
        #max_features=10000,
        max_features=8000,
        #min_df=3,
        min_df=10,
        #max_df=0.9,
        max_df=0.85,
        ngram_range=(1,2),   # VERY IMPORTANT
        stop_words='english',
        token_pattern=r'\b[a-zA-Z]{2,}\b',
        sublinear_tf=True
    )),
    ('select', SelectKBest(chi2, k=3000)),
    ('model', LogisticRegression(
        penalty='elasticnet',
        solver='saga',
        C=10,
        l1_ratio=0.2,
        max_iter=2000,
        random_state=42
    ))
])

pipeline.fit(X_train, y_train)

# Save
with open("TF_IDF_Model.pkl", "wb") as f:
    pickle.dump(pipeline, f)
```

```python
import re
import pandas as pd

def count_urls(text):
    if pd.isna(text):
        return 0
    return len(re.findall(r'(https?://\S+|www\.\S+)', str(text)))

def has_url(text):
    if pd.isna(text):
        return 0
    return int(bool(re.search(r'(https?://\S+|www\.\S+)', str(text))))

def has_ip(text):
    if pd.isna(text):
        return 0
    return int(bool(re.search(r'\b(?:\d{1,3}\.){3}\d{1,3}\b', str(text))))

def count_exclamations(text):
    if pd.isna(text):
        return 0
    return str(text).count("!")

def count_digits(text):
    if pd.isna(text):
        return 0
    return sum(c.isdigit() for c in str(text))

def count_uppercase_words(text):
    if pd.isna(text):
        return 0
    words = str(text).split()
    return sum(word.isupper() and len(word) > 1 for word in words)

def text_length(text):
    if pd.isna(text):
        return 0
    return len(str(text))
```

```python
suspicious_words = [
    "urgent", "verify", "click", "login", "confirm", "password",
    "account", "bank", "security", "update", "limited", "suspend",
    "immediately", "winner", "claim", "free", "prize"
]

urgent_terms = ["urgent", "immediately", "now", "asap", "within 24 hours", "act now"]

def count_suspicious_words(text):
    if pd.isna(text):
        return 0
    text = str(text).lower()
    return sum(word in text for word in suspicious_words)

def urgent_language_score(text):
    if pd.isna(text):
        return 0
    text = str(text).lower()
    return sum(term in text for term in urgent_terms)
```

```python
def extract_features(df):
    df = df.copy()

    # raw text features
    df["num_urls"] = df["text"].apply(count_urls)
    df["has_url"] = df["text"].apply(has_url)
    df["has_ip"] = df["text"].apply(has_ip)
    df["num_exclamations"] = df["text"].apply(count_exclamations)
    df["num_digits"] = df["text"].apply(count_digits)
    df["uppercase_words"] = df["text"].apply(count_uppercase_words)
    df["text_length"] = df["text"].apply(text_length)

    # clean text features
    df["suspicious_word_count"] = df["clean_text"].apply(count_suspicious_words)
    df["urgent_score"] = df["clean_text"].apply(urgent_language_score)

    return df
```

```python
from collections import Counter

def extract_suspicious_words(df, text_col="clean_text"):
    words_list = []

    for text in df[text_col]:
        words = str(text).split()
        for w in words:
            if w in suspicious_words:
                words_list.append(w)

    return Counter(words_list)
```
